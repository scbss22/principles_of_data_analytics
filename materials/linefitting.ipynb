NumPy's polyfit function fits a polynomial of a specified degree to a given set of data points using least squares regression. It returns the polynomial coefficients in descending order, which can then be used for interpolation, approximation, or trend analysis. ## **Data**

https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html

```python
# Create x values.
x = np.linspace(0.0, 10.0, 31)

# Create y values.
y = 3.0 * x
```

```python
# Create a scatterplot.
fig, ax = plt.subplots()

# Plot the data.
ax.scatter(x, y);
```

```python
# Use polyfit to fit a line to the data.
m, c = np.polyfit(x, y, 1)

# Show.
m, c
```

```python
# Create a scatterplot.
fig, ax = plt.subplots()

# Plot the data.
ax.scatter(x, y)

ax.plot(x, m * x + c, color='red');
```

## All together

```python
# Create x values.
x = np.linspace(0.0, 10.0, 31)

# Create y values.
y = 3.0 * x + 1.0 + np.random.normal(0.0, 1.0, size=x.size)

# Use polyfit to fit a line to the data.
m, c = np.polyfit(x, y, 1)

# Show.
m, c

# Create a scatterplot.
fig, ax = plt.subplots()

# Plot the data.
ax.scatter(x, y)

ax.plot(x, m * x + c, color='red')

ax.set_title(f'y = {m:.2f}x + {c:.2f}');
```
